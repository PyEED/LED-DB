## Introduction ##

$\alpha$/$\beta$-hydrolases are enyzmes which fullfill different functions, and can be subsumed into lipases, hydrolases, amidases, proteases and so on. <br/> 
The characteristic structure consists of $\beta$ folds surrounded by two layers of $\alpha$ helical sheets. Every $\alpha$/$\beta$-hydrolases posses <br/> 
an active domain in which nucleophilic aminoacids can be found i.e aspartate, cystein and serine, or catalytic accids i.e aspartate or <br/>
glutamate, and additionally an oxyanion hole is located in the active domain with two different sequences, namely GX and GGGX, in which the catergorization of the <br/>
$\alpha/\beta$-hydrolasis is made. In hydrolasis of the type GX, the first part in the oxyanion hole is formed by the backbon N-H of the X aminioaccid, <br/>
while in the GGGX type it is the N-H of the third G. Slightly or exchanges of one aminoaccid have a significant impact in the functionality of the enzyme, so that for <br/>
for example a lipase can be transfered into an epoxide hydrolase.

```python
# 1. superfamily from the table above beginning with core
bsla = ProteinInfo.from_ncbi("1I6W_B")
lpc = ProteinInfo.from_ncbi("4BZW_B")

# 2. superfamily core lid beta1 to beta 2
bcl = ProteinInfo.from_ncbi("3LIP_A")

# 3. --
rml = ProteinInfo.from_ncbi("4TGL_A")

# 4. --
crl = ProteinInfo.from_ncbi("1TRH_A")

# 5. --
xahd = ProteinInfo.from_ncbi("1EDB_A")
mth = ProteinInfo.from_ncbi("2VF2_B")
cala = ProteinInfo.from_ncbi("2VEO_B")

# 6. --
sceh = ProteinInfo.from_ncbi("4QA9_A")

# 7. -- 
trc = ProteinInfo.from_ncbi("4PSC_A")
pce = ProteinInfo.from_ncbi("3ZWQ_B")

# 8. --
apap = ProteinInfo.from_ncbi("2HU5_B")

# 9. --
rce = ProteinInfo.from_ncbi("1JU4_A")

# 10 --
hpl = ProteinInfo.from_ncbi("1LPB_A")

# 11 -- 
psml = ProteinInfo.from_ncbi("2Z8X_A")

# 12 --
mleh = ProteinInfo.from_ncbi("1CQZ_B")
```
